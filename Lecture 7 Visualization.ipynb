# Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
```

Two plotting styles:
1. simple Matlab style, quick plotting
2. more detailed object oriented plotting

```python
x = np.linspace(0,10,200)
y = np.sin(x)
z = np.cos(x)
```

```python
plt.plot(x,y,'b-',linewidth=10)
```

```python
# fig is a blank piece of canvas, ax is picture on the canvas. There can be more than one ax
fig, ax = plt.subplots()
ax.plot(x,y,'r-',linewidth=2,label='sin(x)',alpha=0.2)
ax.plot(x,z,'b-',linewidth=3,label='cos(x)')
ax.legend(loc='upper right')
ax.set_title('Function plots')
ax.set_yticks([-1,0,1])
plt.show()
```

```python
fig, ax = plt.subplots(2,1,figsize=(10,10))
ax[0].plot(x,y,'r-',linewidth=2,label='sin(x)',alpha=0.2)
ax[1].plot(x,z,'b-',linewidth=3,label='cos(x)')
ax[0].legend(loc='upper right')
ax[0].set_title('Function plots')
ax[0].set_yticks([-1,0,1])
```

```python
countries = ['CAN','MEX','USA']
populations = [36.7,129.2,325.7]
land_area = [3.85, 0.761, 3.79]
```

```python
fig, ax = plt.subplots(2)
ax[0].bar(countries,populations, align='center', color='b')
ax[0].set_title('Populations (in millions)')
ax[1].bar(countries,land_area, align='center', color='r')
ax[1].set_title('Land area (in millions miles squared)')
fig.tight_layout()
```

Generate some random numbers, plot the locaiton of these numbers with random sizes

```python
N = 50
```

```python
rng = np.random.default_rng()
```

```python
x = rng.random(N) # drawn from a uniform distribution in the interval [0,1]
y = rng.random(N)
colors = rng.random(N)
area = np.pi*(35*rng.random(N)**2)
```

```python
fig, ax = plt.subplots()
ax.scatter(x,y,s=area, c=colors, alpha=0.5)
```
